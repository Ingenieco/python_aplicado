```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Aplicando el filtro a partir de las derivadas con Sobel X (deriva respecto a X)

```python
img = cv2.imread("C:\\Jupyter\\pdi\\imagenes\\image.png")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

## Aplicando las derivadas
dx = 1 # derivada de orden 1
dy = 0 # derivada orden 0
tamano_kernel = 3

image_with_sobel_x = cv2.Sobel(img_gray, cv2.CV_64F, dx, dy, tamano_kernel)
```

```python
cv2.imwrite("C:\\Jupyter\\pdi\\imagenes\\imagen_sobel_x.png", image_with_sobel_x)
cv2.imshow("Ventana 1", img)
cv2.imshow("Ventana 2", image_with_sobel_x)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

### Aplicando el filtro a partir de las derivadas con Sobel Y (deriva respecto a Y)

```python
## Aplicando las derivadas
dx = 0 # derivada de orden 1
dy = 1 # derivada orden 0
tamano_kernel = 3

image_with_sobel_y = cv2.Sobel(img_gray, cv2.CV_64F, dx, dy, tamano_kernel)
cv2.imwrite("C:\\Jupyter\\pdi\\imagenes\\imagen_sobel_y.png", image_with_sobel_y)
cv2.imshow("Ventana 1", img)
cv2.imshow("Ventana 2", image_with_sobel_y)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
## Al aplicar la fórmula, se calcula el valor absoluto de las derivadas en x y y.

image_with_sobelx_abs = cv2.convertScaleAbs(image_with_sobel_x)
image_with_sobely_abs = cv2.convertScaleAbs(image_with_sobel_y)

## Luego se hace una unión de las imágenes
joined_images = cv2.addWeighted(image_with_sobelx_abs, 1, image_with_sobely_abs, 1, 0)

cv2.convertScaleAbs(image_with_sobel_y)
plt.figure(figsize = (10,8))
plt.subplot(1,2,1)
plt.title("imagen")
plt.imshow(img, cmap="gray")
plt.xticks([])
plt.yticks([])

plt.subplot(1,2,2)
plt.title("Sobel y")
plt.imshow(joined_images, cmap="gray")
plt.xticks([])
plt.yticks([])

plt.show()
```

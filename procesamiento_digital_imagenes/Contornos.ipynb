```python
'''
Siempre se debe umbralizar

CHAIN_APROX se utiliza cuando hay puntos redundantes en la figura

'''
image_BGR = cv2.imread("C:\\Jupyter\\pdi\\imagenes\\image.png")
image_GRAY = cv2.cvtColor(image_BGR,cv2.COLOR_BGR2GRAY)

valor_maximo = 255
valor_umbral = 127
tipo = cv2.THRESH_BINARY

ret, image_with_threshold = cv2.threshold(image_GRAY, valor_umbral, valor_maximo, tipo)
contours, hierarchy = cv2.findContours(image_with_threshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
#print(contours)
color = (0,255,0)
cv2.drawContours(image_BGR,contours,-1, color,8)
cv2.imshow("imagen umbralizada", image_with_threshold)
cv2.imshow("imagen con contornos", image_BGR)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

### Contornos en video de objetos de color rojo

```python
import cv2
import numpy as np 

cap = cv2.VideoCapture("C:\\Jupyter\\pdi\\imagenes\\circles.avi")

Redlower1 = np.array([0,80,80], np.uint8)
Redupper1 = np.array([8,255,255], np.uint8)

Redlower2 = np.array([175,80,80], np.uint8)
Redupper2 = np.array([179,255,255], np.uint8)

while cap.isOpened():
	ret, frame = cap.read()
	frame_hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
	mascara1 = cv2.inRange(frame_hsv, Redlower1, Redupper1)
	mascara2 = cv2.inRange(frame_hsv, Redlower2, Redupper2)
	mascara = cv2.add(mascara1, mascara2)
	contours, hierarchy = cv2.findContours(mascara,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
	color = (0,255,0)
	cv2.drawContours(frame,contours,-1,color,2)
	#filtro_img = cv2.bitwise_and(frame,frame,mask = mascara)
	if ret == True:
		cv2.imshow("Video", frame)
		#cv2.imshow("mascara",mascara)
		#cv2.imshow("filtro_img",filtro_img)
		#print(frame)
		if cv2.waitKey(100) == 97: # desactivar el video con la letra a del teclado
			break
	else:
		break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

'''
No olvidar que aquí se estima es el área del CONTORNO NO EL ÁREA DE LA IMAGEN. Además, el perímetro y el área están en unidades de píxeles, pero se pueden convertir a milímetros, etc.
'''



image_BGR = cv2.imread("C:\\Jupyter\\pdi\\imagenes\\imagen_cuadrado.jpg")
image_GRAY = cv2.cvtColor(image_BGR, cv2.COLOR_BGR2GRAY)

valor_maximo = 255
valor_umbral = 127
tipo = cv2.THRESH_BINARY

retval, image_with_threshold = cv2.threshold(image_GRAY, valor_umbral, valor_maximo,tipo)
print(f"umbral {retval} y la imagen umbralizada {image_with_threshold}")
contours, hierarchy = cv2.findContours(image_with_threshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contour = contours[0]
moments = cv2.moments(contour)
print("Momentos 00 (área): ", moments['m00'])

area = cv2.contourArea(contour)
print(f"Área {area}")

p = cv2.arcLength(contour, True)
print(f"Perímetro {p}")
```

## Cuadro delimitador

```python
import cv2

image_BGR = cv2.imread("C:\\Jupyter\\pdi\\imagenes\\imagen_para_cuadrodelimitador.png")
image_GRAY = cv2.cvtColor(image_BGR,cv2.COLOR_BGR2GRAY)

valor_maximo = 255
valor_umbral = 127
tipo = cv2.THRESH_BINARY

ret,image_with_threshold = cv2.threshold(image_GRAY,valor_umbral,valor_maximo,tipo)

contours, hierarchy = cv2.findContours(image_with_threshold,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
contour = contours[0]

x,y,w,h = cv2.boundingRect(contour) # w, h es el ancho y alto del cuadro delimitador. x, y son las puntos de la esquina superior izquierda. por tanto a partir de x, y se suma el ancho y el alto para delimitar 

color = (0,255,0)
cv2.rectangle(image_BGR,(x,y),(x+w,y+h),color,2)

cv2.imshow("imagen con contornos",image_BGR)
cv2.waitKey(100000)
cv2.destroyAllWindows()
```
